```python
import pandas as pd 
import numpy as np
import matplotlib
%matplotlib inline
import seaborn as sns 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import random
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.width', 5000)
data = pd.read_csv('C:\\Sohan\\ML_Acmegrade\\ML Day13 5\\pacific.csv')
data.head()
```

```python
#dependent var(Y)
data.groupby('Status').agg({'Status': np.size})
```

```python
#label encoding with cat.codes ; 
data.Status = pd.Categorical(data.Status)
data['Status'] = data.Status.cat.codes
data.head()
```

```python
data.groupby('Status').agg({'Status': np.size})
```

```python
random.seed(2)
pred_columns = data[:]
pred_columns.drop(['Status'],axis=1,inplace=True)#coz thats Y column
pred_columns.drop(['Event'],axis=1,inplace=True)#rest of these are unwanted data
pred_columns.drop(['Latitude'],axis=1,inplace=True)
pred_columns.drop(['Longitude'],axis=1,inplace=True)
pred_columns.drop(['ID'],axis=1,inplace=True)
pred_columns.drop(['Name'],axis=1,inplace=True)
pred_columns.drop(['Date'],axis=1,inplace=True)
pred_columns.drop(['Time'],axis=1,inplace=True)
prediction_var = pred_columns.columns
print(list(prediction_var))# my X column
```

```python
from sklearn.model_selection  import train_test_split
train, test = train_test_split(data, test_size = 0.3)
print(train.shape)
print(test.shape)
#here we did train to split for the whole data
```

```python
train
```

```python
#hence we split now for X n Y
train_x = train[prediction_var]
train_y= train['Status']  
print(list(train.columns))
print(train_x.shape)
print(train_y.shape)
```

```python
test_x= test[prediction_var] 
test_y =test['Status']
print(list(test.columns))
print(test_x.shape)
print(test_y.shape)
```

```python
#Logistic Regg    
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(train_x, train_y)
log_pred = logmodel.predict(data[prediction_var])
print (log_pred)
print (pd.DataFrame(log_pred).groupby(0).agg({0: np.size}))
print('\n Accuracy Score')
print (accuracy_score (data['Status'],log_pred))
print('\nClassification Report')
print(classification_report (data['Status'],log_pred))
print('Confusion Matrix')
print (confusion_matrix (data['Status'],log_pred))
#acc = 0.76
#as we can see only label encoded as 3,4,9,10 can be predicted since once these columns has sufficient data 
#length of confusion matrix is total no. of distinct values in Y(here its 11 so 11x11 confusion matrix is created)
```

```python
data.groupby('Status').agg({'Status': np.size})#we can cross check confusion matrix with this
```

```python
# Model KNN  with Euclidian distance  ie P =2 
from sklearn.neighbors import KNeighborsClassifier
KNN1 = KNeighborsClassifier(n_neighbors =100 , metric = 'minkowski', p= 2)
KNN1.fit(train_x,train_y)
KNN1_pred = KNN1.predict(data[prediction_var])
print (KNN1_pred)
print (pd.DataFrame(KNN1_pred).groupby(0).agg({0: np.size}))
print('\n Accuracy Score')
print (accuracy_score (data['Status'],KNN1_pred))
print('\nClassification Report')
print(classification_report (data['Status'],KNN1_pred))
print('Confusion Matrix')
print (confusion_matrix (data['Status'],KNN1_pred))
#acc = 0.95
```
